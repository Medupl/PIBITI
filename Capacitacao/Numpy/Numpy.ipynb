### Introdução a Numpy
---
Oferece suporte para arrays e matrizes multidimensionais, além de fornecer uma coleção de funções matemáticas para operar sobre esses arrays. É amplamente utilizada em ciência de dados, aprendizado de máquina, álgebra linear, estatísticas, entre outros.
+ Algumas funcionalidades do NumPy:
    - Operações com Arrays;
    - Operações vetorizadas;
    - Operaçoes matemáticas;
    - Integrações com outras bibliotecas, pandas e Matplotlib.

```python
# importando as bibliotecas
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Alguns exemplos básicos:

# Criando matrizes
matriz = np.array([[1, 2, 3], [4, 5, 6]])           # Criando matrizes manualmente, nesse caso 2x3.
print("Primeira Matriz:\n",matriz)
zeros = np.zeros((3, 3))                            # Matriz automatica 3x3 de zeros
uns = np.ones((3, 2))                               # Matriz automatica 2x4 de uns
arange = np.arange(0, 10, 2)                        # Array de 0 a 10 com intervalo 2, sem inclusão do 10.
linspace = np.linspace(0, 1, 5)                     # 5 valores igualmente espaçados entre 0 e 1

# Operações com matrizes.
matriz = np.dot(matriz,uns)                         # Realizando multiplicação de matrizes.
raiz_quadrada = np.sqrt(matriz)                     # Raiz quadrada dos elementos da matriz.
log = np.log(matriz)                                # Logaritmo dos elementos da matriz.
matriz_T = matriz.T                                 # Realizando transpota da matriz.

print("Multiplicação das Matrizes:\n",matriz)
print("Transposta da Matriz:\n",matriz_T)
print("Forma da Matriz:\n",matriz.shape)
print("Tipo de dados:\n",matriz.dtype)
print("Número total de elementos:\n",matriz.size)
```

```python
# Funções Estatisticas:

media = np.mean(matriz)                             # Achando a média da matriz.
soma = np.sum(matriz)                               # Somando todos os valores da matriz.
minimo = np.min(matriz)                             # Encontrando o valor minimo.
maximo = np.max(matriz)                             # Encontrando o valor maximo.
desvio_padrao = np.std(matriz)                      # Encontrando o desvio padrão da matriz.

print("Media da Matriz:\n",media)
print("desvio da Matriz:\n",desvio_padrao)
```

### Matplolib
---
O Matplotlib permite criar uma ampla variedade de gráficos, personalizar o estilo, e ajustar o tamanho e layout dos gráficos. O $pyplot$ é um módulo dentro do Matplotlib que oferece uma interface de estilo similar ao MATLAB para criar gráficos de forma fácil e rápida.

```python
# Gerar dados
t = np.linspace(0, 3 * np.pi, 100)
y = np.sin(t)
z = np.cos(t)
categorias = ['A', 'B', 'C', 'D']
valores = [3, 7, 2, 5]

# Configurando o gráfico.
plt.figure(figsize=(12, 2))
plt.subplots_adjust(
    left=0.1,
    right=0.9,
    top=0.9,
    wspace=0.4,
    hspace=0.4
)

# Gráfico da função seno
plt.subplot(1, 3, 1)
plt.plot(t, y)
plt.grid()
plt.title('Função Seno')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')

# Gráfico da função cosseno
plt.subplot(1, 3, 2)
plt.plot(t, z)
plt.grid()
plt.title('Função Cosseno')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')

# Gráfico de pizza
plt.subplot(1, 3, 3)
plt.pie(valores, labels=categorias, autopct='%1.2f%%', startangle=90)
plt.title('Gráfico de Pizza')

plt.show()
```

### Gerando amostra aleatórias com Numpy
---
Retomando o estudo sobre Numpy, usaremos o novo gerador de números aleatórios da biblioteca Numpy, o $default$_$rng$, que pode ser usado fazendo o seguinte import: `from numpy.random import default_rng` .
+ Segue  alguns exemplos abaixo:

```python
from numpy.random import default_rng
from time import process_time

# Exemplo básico do uso da biblioteca.
rng = default_rng()

aleatorio = rng.integers(20, size=(3,4))
# Gerando listas aleatórias de números ineiros.
list1 = list(rng.integers(10, 100, 10000000))
list2 = list(rng.integers(10, 100, 10000000))

print(aleatorio)
print("--------------")
print("Lista 1: ", type(list1))
print("Lista 2: ", type(list2))
```

> Caso tente, fazer uma multiplicação entre as listas usando: `x = list1 * list2`, isso resultrá em um erro. O operador * entre duas listas Python não faz uma multiplicação elemento a elemento. O código está tentando multiplicar duas listas diretamente, o que não é permitido. Para realizar uma multiplicação elemento a elemento entre duas listas de números, você precisaria usar NumPy ou um loop.

```python
# Iniciando uma lista vazia
x=[]
# Medindo o tempo do processamento a ser realizado.
t1 = process_time()
for i in range(len(list1)):
    x.append(list1[i] * list2[i])
t2 = process_time()
print(t1)
print(t2-t1)
```

```python
# Realizando o processo acima de outra forma.
# Gerando dessa vez dois arrys.
x = rng.integers(10, 100, 10000000)
y = rng.integers(10, 100, 10000000)
print(type(x))
# Medindo o tempo do processo abaixo.
t1 = process_time()
z = x * y
t2 = process_time()
print("O tempo total de execução da multiplicação:",t2-t1,"s")
```

### Exemplo 2
---
 Este é um exemplo avançado de processamento de informações e cálculos estáticos usando NumPy, focado em análise estatística de um grande conjunto de dados. O exemplo simula o processamento de dados de uma pesquisa, analisa a distribuição de dados, calcula estatísticas descritivas e realiza um teste de hipóteses (teste z).

 > Vamos simular dados de uma pesquisa onde medimos a altura de uma amostra de 100.000 pessoas. Usaremos NumPy para realizar cálculos estatísticos avançados, como:
  + Distribuições de probabilidade (distribuição normal).
  + Cálculos estatísticos básicos (média, mediana, desvio padrão).
  + Análise de outliers (dados fora de 2 desvios padrão da média).
  + Teste de hipóteses (teste z) para comparar a média da amostra com a média da população.


```python
# Importando biblioteca para tratar dados estatísticos
from scipy import stats

# 1. Simulação de Dados: Altura (em cm) de 100.000 pessoas, distribuição normal.
# Regra Empírica: 68% dos valores estão a 1 dp (𝜇 ± 𝜎); 95% dos valores estão a 2 dp (𝜇 ± 2𝜎); 99.7% dos valores estão a 3 dp (𝜇 ± 3𝜎).
# Vamos definir uma média dos valores: 170 cm e desvio padrão: 10 cm
# Vamos definir também uma semente para gerar os mesmos dados aleatórios sempre que rodarmos a simulação:
np.random.seed(40)
alturas = np.random.normal(loc=170, scale=10, size=100000)

# 2. Cálculos estatísticos básicos
media = np.mean(alturas)
mediana = np.median(alturas)
desvio_padrao = np.std(alturas)

print(f"Média: {media:.3f} cm")
print(f"Mediana: {mediana:.3f} cm")
print(f"Desvio Padrão: {desvio_padrao:.3f} cm")
```

```python
# 3. Identificação de Outliers (Alturas fora de 2 desvios padrão da média)
# Usando a regra empírica, 95% dos valores estão a 2 dp (𝜇 ± 2𝜎), então usar dois desvios como parâmetro é uma boa escolha.

# Definindo os valores extremos que serão usados para comparação.
limite_inferior = media - (2 * desvio_padrao)
limite_superior = media + (2 * desvio_padrao)

# Indentificação dos valores de altura que estão abaixo ou acima dos limites.
outliers = alturas[(alturas < limite_inferior) | (alturas > limite_superior)]

print(f"Quantidade de Outliers: {len(outliers)}")
print(f"Outliers: {outliers}")
```

```python
# Importando biblioteca que cria gráficos de alta qualidade.
import seaborn as sns

# Configurando os gráficos que serão apresentados:
fig, axis = plt.subplots(1, 2, figsize=(15, 3))

# 4. Histograma e Curva de Densidade (Distribuição de Alturas)
axis[0].hist(alturas, bins=50, density=True, alpha=0.6, color='g', label='Alturas')
axis[0].set_title("Histograma das alturas")
axis[0].set_xlabel("Altura (cm)")
axis[0].set_ylabel("Densidade")
axis[0].legend()

# Adicionando a curva de densidade
sns.kdeplot(alturas, ax = axis[1], color='blue', label='Curva de Densidade', linewidth=2)
axis[1].set_title("Curva de Densidade das alturas")
axis[1].set_xlabel("Altura (cm)")
axis[1].set_ylabel("Densidade")
axis[1].legend()

# Curva de densidade (função de distribuição normal)
xmin, xmax = plt.xlim()                                   # Calcula os valores máximo e minimos de x
x = np.linspace(xmin, xmax, 100)                          # Cria um array igualmente espaçados.
density = stats.norm.pdf(x, media, desvio_padrao)         # Calcula os valores da função densidade mormal
axis[0].plot(x, density, 'r', linewidth=2, label='Distribuição Normal', alpha = 0.5)
axis[0].legend()

# Mostrar os gráficos.
plt.show()
```

```python
# 5. Teste de Hipóteses: Teste Z
# Hipótese nula (H0): A média das alturas é igual à média da população (170 cm)
# Hipótese alternativa (H1): A média das alturas não é igual à média da população

# Parâmetros da população (população é conhecida neste caso)
media_populacao = 170
desvio_padrao_populacao = 10

# Tamanho da amostra
n = len(alturas)

# Estatística Z (Teste Z)
z_statistica = (media - media_populacao) / (desvio_padrao_populacao / np.sqrt(n))

# Valor-p para o teste Z
p_valor = 2 * (1 - stats.norm.cdf(np.abs(z_statistica)))

print(f"Estatística Z: {z_statistica:.2f}")
print(f"Valor-p: {p_valor:.5f}")

# Interpretação do Resultado
significancia = 0.05  # Nível de significância de 5%
if p_valor < significancia:
    print("Rejeitamos a hipótese nula. A média da amostra é significativamente diferente da média da população.")
else:
    print("Não rejeitamos a hipótese nula. A média da amostra não é significativamente diferente da média da população.")
```
