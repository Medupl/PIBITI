#### Commit
---
Ao inciar a criação de arquivos ou pastas dentro do nosso repositório, precisamos dar um $commit$ em alguma parte específica do código ou no repositório. Um commit é importante para ter controle de funções, backups, facilita a comunicaão com outros desenvolvedores e permite um melhor controle de seu projeto.

+ O primeiro passo é darmos um $git$ $status$ onde o Git verifica se houve alguma alteração. Neste passo pode surgir 3 situações: $untracked, modified$ e $staged$
  - $untracked:$ Quando seu arquivo é novo e ainda não esta sendo visto pelo Git. Para resolver este caso teremos que fazer um $git$ $add$ ., assim o Git já terá controle que estamos com um arquivo novo.
  
    <center><img src='https://github.com/Medupl/CoppeliaSim/blob/main/images/Git_GitHub/gitStatus.png?raw=true'/></center>

  - $Modified:$ Quando fizemos alguma alteração em um arquivo, mas que já foi adicionado anteriormente. $deleted:$ Algum arquivo que foi deletado do projeto e que precisa ser fazer um commit para deixar registrado no projeto.

    <center><img src='https://github.com/Medupl/CoppeliaSim/blob/main/images/Git_GitHub/gitModified.png?raw=true'/></center>

  - $staged:$ Os arquivos neste estagio já estão prontos para serem commitados, eles ficam neste estágio logo após serem adicionados pelo $git$ $add$. 

    <center><img src='https://github.com/Medupl/CoppeliaSim/blob/main/images/Git_GitHub/gitStaged.png?raw=true'/></center>

#### Comunicação com o GitHub
---
+ O proximo passao é adicionar os arquivos e atualizações no GitHub. Assim, se faz necessário criar um repositório no GitHub, o nome desse repositório deve ser idêntico ao nome da pasta local que foi criada no computador. Isso garante uma organização consistente entre os arquivos locais e o repositório remoto. Após a criação do repositório, é preciso configurar a comunicação entre o repositório local e o GitHub. Para isso, utilizamos o terminal Git Bash, onde realizamos o acesso remoto do repositório.

  - $Push:$ Fazer um push no Git significa enviar as alterações feitas em um repositório local para um repositório remoto. O push permite compartilhar o projeto com outros colaboradores ou armazenar em um servidor remoto.

### Manipulações no Git
---
  + $Git$ $log:$ Exibe todas as informações dos commits realizados, como autor, data, mensagem e para qual commit o ponteiro do repositório $HEAD$ está apontando.
  
    <center><img src="https://github.com/Medupl/CoppeliaSim/blob/main/images/Git_GitHub/gitLog.png?raw=true"/></center>
  
<center>Ao utilizar o "--oneline -5", estamos mostrando somente as cinco últimas atualizações.</center>
  
  > $Graph:$ Para exibir o gráfico dos commits, devido em projetos precisarmos trabalhar com mais branchs.
  Usamos o seguinte comando:    <center>`git log --oneline --graph`</center>
  
+ $Gitignore:$ Para manter o repositório limpo, organizado e seguro, criamos um arquivo $.gitignore$. Neste arquivo especificamos quais arquivos ou diretórios devem ser ignorados pelo Git, isso significa que esses arquivos não será adicionado no repositório do GitHub.
  
  <center><img src="https://github.com/Medupl/CoppeliaSim/blob/main/images/Git_GitHub/gitIgnore.png?raw=true"/></center>

> Neste exemplo, após a criação do arquivo .txt, adicionamos ele no $.gitignore$. Ao ser infomado neste arquivo ele não será reconhecido pelo Git e não será incluindo no repositório.

+ $Git$ $reset:$ Este comando é usado para desfazer alterações no Git. Ele pode modificar commits, mover ponteiros e alterar o $staged$. Este comando possui três modelos:
  - $soft:$ Move o ponteiro para o commit desejado, mas mantém as alterações no $staged$ e no diretório.
   Código:<center> `git reset --soft HEAD~1`</center>
  - $mixed:$ Move o ponteiro para o commit desejado, remove as alterações no $staged$, mas mantém as altereções no diretório, terá que adicionar novamente ao $staged$ caso deseje fazer um novo commit.
   Código:<center> `git reset --mixed HEAD~1`</center>
  - $hard:$ Move o ponteiro para o commit desejado, remove as alterações no $staged$, e descarta as altereções no diretório, qualquer informação não commitada será removida. Usar com cuidado.
   Código:<center> `git reset --hard HEAD~1`</center>
> $Git$ $revert:$ Esse comando cria um novo commit que desfaz as alterações introduzidas por um commit anterior. Esse comando não altera o histórico de commits como o git reset faz, mas adiciona um novo commit que inverte as mudanças do commit alvo.

### Branchs
---
> Branchs são usadas para criar diferentes versões do projeto em paralelo. Permitindo trabalhar em funções, correções ou ajustes de forma isolada sem afetar a versão principal do projeto.
  - $Branch$ $principal:$ Geralmente chamada de $main$, que contém a versão estável e principal do código. Onde ficam as funcionalidades principais do projeto.
  - $Branch$ $secundária:$  Usada para trabalhar em funcionalidades novas, correções ou ajustes de forma isolada.

    - Criando uma nova branch e realizando a verificação da lista de branchs:
      <center><img src="https://github.com/Medupl/CoppeliaSim/blob/main/images/Git_GitHub/gitBranch.png?raw=true"/></center>

    - Mudando para a branch secundária onde podemos realizar novas mudanças sem afetar a main. Assim que criamos a nova branch ela automaticamente receber todas as informações da principal. A partir deste momento para ficar com o ponteiro na mesma direção da principal, será preciso fazer uma merge.
  
      <center><img src="https://github.com/Medupl/CoppeliaSim/blob/main/images/Git_GitHub/gitCheckout.png?raw=true"/></center>
  
  - $Rebase:$ é usado no Git para mover ou combinar commits de uma branch sobre outra. Ele reaplica os commits de uma branch em cima de outra. Isso resulta em um histórico linear, sem commits de merge, o que pode ser mais fácil de entender e visualizar. Preservando também a relação pai/filho dos commits.
  
    <center><img src="https://github.com/Medupl/CoppeliaSim/blob/main/images/Git_GitHub/gitRebase.png?raw=true"/></center>

  > Nesse caso, o Git "reaplicará" os commits da branch $functions-new$ em cima dos commits mais recentes da branch main, como se você tivesse começado a trabalhar com base na última versão da branch main. Para essas alterações irem para a $main$, podemos fazer um $merge$ ou $rebase$, porém não terá o commit de merge e mantendo a linearidade.

  - $Merge:$ Quando você faz um merge, ele combina os commits de duas branches e cria um commit de merge que preserva o histórico paralelo das duas branches. O merge é recomendado para equipes colaborativas, pois ele preserva a sequência de commits de todas as branches, sem modificar o histórico original. Porém não preserva a relação pai/filho.
  
    <center><img src="https://github.com/Medupl/CoppeliaSim/blob/main/images/Git_GitHub/git_merge.png?raw=true"/></center>

  > Ambos têm seus prós e contras, e a escolha depende de como você prefere gerenciar o histórico do seu projeto e da colaboração com outros desenvolvedores.

### Fork e Pull Request
---
- O $Fork$ é uma cópia de um repositório que fica sob a sua conta, mas que mantém um vínculo com o repositório original. O processo de fazer um fork é frequentemente usado quando você quer colaborar em um projeto no qual você não tem permissões de escrita diretamente.
  O fork permite que você faça alterações no repositório de outra pessoa sem afetar o projeto original.
  Quando terminar, você pode propor que suas alterações sejam incorporadas ao projeto original através de um pull request.

  - Depois de fazer o fork, clone o repositório para sua máquina local para poder trabalhar nele. Sempre crie uma nova branch para fazer suas alterações. Isso mantém um repositório organizado.

- $Pull$ $request:$ é uma solicitação para que as alterações que você fez no seu fork (ou em uma branch de um repositório) sejam revisadas e incorporadas ao repositório original. Ele é usado para iniciar a revisão de código e a discussão das mudanças feitas.
  - Após fazer alguma alteração construtiva ao projeto, você pode criar um pull request para que o repositório original (onde você fez o fork) possa revisar suas mudanças.

    - Vá até o seu fork no GitHub.
    - Clique no botão Compare & pull request.
    - Preencha os detalhes do pull request, incluindo uma descrição clara do que suas alterações fazem.
    - Envie o pull request.
